Problem #1 : Part A

```python
endpoint1 = '/symbols'
url1 = base_url + endpoint1
r1 = requests.get(url1, headers=headers)
# print(r1.ok)
# print(r1.url)
symbols = r1.json()['symbols']
print(symbols)
```

```python
codes = ['MXN', 'BRL', 'HUF', 'KRW', 'ZAR', 'NZD', 'AED', 'USD']

for code in codes:
    if code in symbols:
        print(f'{code}: {symbols[code]}')
```

Part B

```python
endpoint2 = '/timeseries'
url2 = base_url + endpoint2
param2 = {
  "base": "EUR",
  "start_date": "2023-08-01",
  "end_date": "2023-10-08",
  "symbols": ','.join(codes)
}
r2 = requests.get(url2, headers=headers, params=param2)
print(r2.ok)
print(r2.url)
```

```python
print(r2.text)
```

```python
series = r2.json()['rates']
print(series)
```

```python
df = pd.DataFrame.from_dict(series, orient='index')
df
```

```python
correlations = df.corr()['USD']
correlations = correlations.drop('USD')
print(correlations)

highest_positive = correlations.idxmax(), correlations.max()
highest_negative = correlations.idxmin(), correlations.min()

print("Highest positive correlation with USD:", highest_positive)
print("Highest negative correlation with USD:", highest_negative)
```

```python
df.to_csv('currency_exchange.csv')
```

```python
with open('weather_key.txt', 'r') as infile:
    apikey3 = infile.readline().strip()

headers3= {
    "?access_key": apikey3
}
base_url3 = "http://api.weatherstack.com/"
```

```python
state_capitals_and_others = [
    'Montgomery', 'Juneau', 'Phoenix', 'Little Rock', 'Sacramento', 'Denver', 'Hartford', 'Dover', 
    'Tallahassee', 'Atlanta', 'Honolulu', 'Boise', 'Springfield', 'Indianapolis', 'Des Moines', 
    'Topeka', 'Frankfort', 'Baton Rouge', 'Augusta', 'Annapolis', 'Boston', 'Lansing', 'St. Paul', 
    'Jackson', 'Jefferson City', 'Helena', 'Lincoln', 'Carson City', 'Concord', 'Trenton', 'Santa Fe', 
    'Albany', 'Raleigh', 'Bismarck', 'Columbus', 'Oklahoma City', 'Salem', 'Harrisburg', 'Providence', 
    'Columbia', 'Pierre', 'Nashville', 'Austin', 'Salt Lake City', 'Montpelier', 'Richmond', 'Olympia', 
    'Charleston', 'Madison', 'Cheyenne', 'New York', 'Provo', 'Orem', 'Mesa'
]
```

```python
endpoint3 = f'current'
url3 = base_url3 + endpoint3

# Empty list to store the data
weather_data = []

for city in state_capitals_and_others:
    param3 = {
        "access_key": apikey3,
        "query": city,
        "units": "f"  # Fahrenheit
    }
    r3 = requests.get(url3, params=param3)
    data = r3.json()
    
    # Extract the needed information
    if 'current' in data:
        weather_data.append({
            'City': city,
            'Temperature': data['current']['temperature'],
            'Humidity': data['current']['humidity'],
            'Wind Speed': data['current']['wind_speed']
        })
    else:
        print(f"Error retrieving data for {city}: {data.get('error', 'No error message available')}")

# Convert the list into a DataFrame
df = pd.DataFrame(weather_data)

print(df)
```

```python
df.to_csv('my_api_data.csv')
```
